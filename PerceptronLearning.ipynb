```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,1])
w = np.zeros(2)
b = 0
lr = 0.1
errors = []
epochs = 400
def sigmoid(x):
    return 1/(1+np.exp(-x))
for _ in range(epochs):
    total_error = 0
    for i in range(len(x)):
        z = np.dot(x[i],w)+b
        y_pred = sigmoid(z)
        error = y[i]-y_pred
        w += lr*error*x[i]
        b += lr*error
        total_error += abs(error)
    errors.append(total_error)
print(w, b)
cnt=0
for i in range(len(x)):
    z = np.dot(x[i],w)+b
    y_pred = 1 if sigmoid(z)>=0.5 else 0
    print(x[i], "->", y_pred)
    if (y_pred==y[i]): cnt+=1
for i in range(len(x)):
    plt.scatter(x[i][0],x[i][1],c="green" if y[i]==1 else "red")
x1 = np.linspace(-0.4,1,10)
x2 = -(w[0]*x1 + b)/w[1]
plt.plot(x1, x2, 'b-')
plt.title('OR Gate with Sigmoid Perceptron')
plt.xlabel('X1'); plt.ylabel('X2')
plt.show()
plt.plot(errors)
plt.title('Error vs Epoch')
plt.xlabel('Epoch'); plt.ylabel('Total Error')
plt.show()
print(errors[-1])
print(f"{cnt/len(x)*100}%")
```
