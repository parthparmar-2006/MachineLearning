```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../datasets/Income Data.csv')
df.sample(3)
```

```python
plt.scatter(df['income'], df['age'])
```

```python
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11), wcss)
```

```python
km = KMeans(n_clusters=4)
y_means = km.fit_predict(df)
y_means
```

```python
centroids = km.cluster_centers_
centroids
```

```python
plt.scatter(df.iloc[y_means == 0, 0], df.iloc[y_means == 0, 1], color='blue')
plt.scatter(df.iloc[y_means == 1, 0], df.iloc[y_means == 1, 1], color='red')
plt.scatter(df.iloc[y_means == 2, 0], df.iloc[y_means == 2, 1], color='green')
plt.scatter(df.iloc[y_means == 3, 0], df.iloc[y_means == 3, 1], color='yellow')
plt.scatter(centroids[:,0], centroids[:,1], marker='X', color='black')
plt.show()
```

```python
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled.sample(3)
```

```python
plt.scatter(df_scaled['income'], df_scaled['age'])
```

```python
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df_scaled)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11), wcss)
```

```python
km = KMeans(n_clusters=3)
y_means = km.fit_predict(df_scaled)
y_means
```

```python
centroids = km.cluster_centers_
centroids
```

```python
plt.scatter(df_scaled.iloc[y_means == 0, 0], df_scaled.iloc[y_means == 0, 1], color='blue')
plt.scatter(df_scaled.iloc[y_means == 1, 0], df_scaled.iloc[y_means == 1, 1], color='red')
plt.scatter(df_scaled.iloc[y_means == 2, 0], df_scaled.iloc[y_means == 2, 1], color='green')
plt.scatter(centroids[:,0], centroids[:,1], marker='X', color='black')
plt.show()
```
